# Iris Classification with Keras

#### Import Library and Load Dataset.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['species'],axis=1).values
```

#### Generate dummies label to generate species columns as trained label

```python
y = pd.get_dummies(df['species']).values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

#### 4 Columns 120 Rows Data.

```python
X_train.shape
```

```python
y.shape
```

As this neural network model classifies multiple label it has 3 dimensional shape (Categorical)

#### Scale Dataset since neural networks require scaling numerical dataset to fast convergence

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

#### We only fit on train data, to avoid prior asumption.

```python
scaled_x = scaler.transform(X_train)
```

```python
scaled_xx = scaler.transform(X_test)
```

#### Initiate Keras Models

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

#### Model Description

```python
model = Sequential()
model.add(Dense(8,input_dim=4,activation='relu'))
model.add(Dense(8,input_dim=4,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x=scaled_x,y=y_train,epochs=600)
```

#### Report Probability of each classes

```python
model.predict(scaled_xx)
```

#### Predict classes directly

easiest way is using in built function directly. and predict its classes.

```python
model.predict_classes(scaled_xx)
```

#### Measuring accuracy

Our validation data.

```python
y_test.argmax(axis=1)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_test.argmax(axis=1),model.predict_classes(scaled_xx))
```

```python
print(classification_report(y_test.argmax(axis=1),model.predict_classes(scaled_xx)))
```

```python
accuracy_score(y_test.argmax(axis=1),model.predict_classes(scaled_xx))
```
